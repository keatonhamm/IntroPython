```python
fig, ax1 = plt.subplots()

ax1.plot(x,x**2,lw=2,color="blue")
ax1.set_ylabel("area ($m^2$)", fontsize = 18, color = "blue")
for label in ax1.get_yticklabels():
    label.set_color("blue")

ax2 = ax1.twinx()
ax2.plot(x,x**3,lw=2,color="red")
ax2.set_ylabel("volume ($m^3$)", fontsize=20,color="red")
for label in ax2.get_yticklabels():
    label.set_color("red")
```

```python
n = np.array([0,1,2,3,4,5])
xx = np.arange(-1,1,.1)
fig, axes = plt.subplots(1,4,figsize=(12,4))

axes[0].scatter(xx,xx+.1*np.random.randn(len(xx)))
axes[0].set_title("Scatter Plot")

# Step Plot
axes[1].step(n,n**2,lw=4)
axes[1].set_title("Step Plot")

# Bar plot
axes[2].bar(n,n**2,align="center",width=.5,alpha=.5)
axes[2].set_title("Bar Plot")

# Fill_between
#x2 = np.linspace(0,2,100)
axes[3].fill_between(xx,xx**2,xx**3,color="green",alpha=.5)
axes[3].set_title("fill_between")

plt.show()
```

```python
# Polar coordinates
fig = plt.figure()
ax = fig.add_axes([0,0,1,1],polar=True)
t = np.linspace(0,2*np.pi,100)
ax.plot(t,np.cos(t),color="blue",lw=3)
plt.show()
```

```python
alpha = .7
phi_ext = 2*np.pi*.5

def flux_qubit_potential(phi_m,phi_p):
    return 2+alpha-2*np.cos(phi_p)*np.cos(phi_m)-alpha*np.cos(phi_ext-2*phi_p)
```

```python
phi_m = np.linspace(0,2*np.pi,100)
phi_p = np.linspace(0,2*np.pi,100)
X,Y = np.meshgrid(phi_m,phi_p)
Z = flux_qubit_potential(X,Y)

print(Z.shape)
```

```python
# heatmap
fig, ax = plt.subplots()

color = matplotlib.cm.Spectral
#color = matplotlib.cm.coolwarm

p = ax.pcolor(X,Y,Z, cmap = color)
cb = fig.colorbar(p, ax=ax)
```

```python
# contour plot
fig, ax = plt.subplots()

contour = ax.contour(Z, cmap = color)
cb = fig.colorbar(contour,ax=ax)
```

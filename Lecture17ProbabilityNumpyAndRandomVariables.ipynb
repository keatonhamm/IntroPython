```python
# Generate a random low-rank data matrix

num_cols = 6
num_rows = 6
rank = 2

A = np.random.randn(num_rows,rank)
B = np.random.randn(rank,num_cols)
X = A@B
print(X)

# Randomly select columns and rows
cols = np.sort(np.random.choice(num_cols,rank+2,replace=False))
print(cols)
rows = np.sort(np.random.choice(num_rows,rank+1,replace=False))
print(rows)

# Form submatrices
C = X[:,cols]
print(C)
R = X[rows,:]
print(R)
U = C[rows,:]
print(U)

print("Error", np.linalg.norm(X - C@np.linalg.pinv(U)@R))
print("Difference", X - C@np.linalg.pinv(U)@R)
```

```python
x = np.linspace(0,2*pi,1000) # xmin, xmax, number of points (there are num_point-1 intervals evenly spaced)
x2 = np.arange(0,4,.1) # xmin, xmax, length of intervals
y1 = np.sin(x)
y2 = np.cos(x2)
plt.plot(x,y1)
plt.plot(x2,y2,'r')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title("Graphs of sine and cosine", fontsize = 24,pad=30)
plt.axis([0,2*pi,-1,1]) # [xmin, xmax, ymin, ymax] 
plt.legend(['y = sin(x)', 'y = cos(x)'])
plt.show()
```

1. Make a deck of cards as a numpy array (value, suit)
1. Make a function that draws a hand from the deck of the form
def draw_hand(deck, num_cards = 5):
1. Make a function that randomly deals num_players hands of size num_cards from your deck 
def deal_hands(deck,num_players=9,num_cards=5):

Note: you may either use np.random.choice with replace=False to draw cards, OR you can use np.random.shuffle to first shuffle your deck, and then just draw hands off the top

```python
import numpy as np

values = list(range(2,11))+["Jack","Queen","King","Ace"]
suits = ["Spades","Clubs","Diamonds","Hearts"]

def make_deck():
    deck = list()
    for j in suits:
        for i in values:
            deck.append((i,j))
    deck = np.asarray(deck)
    return deck


deck = make_deck()
print(deck)
```
